```python
station_date = df.loc[(df.index.year == 2019), ]
pd.pivot_table(station_date, values = "ENTRIES", index = "STATION", aggfunc = np.mean).sort_values(by = "ENTRIES", ascending = False).head(5)
```

```python
grouped_date = df.groupby('DATE')
grouped_station = df.groupby("STATION")
```

```python
grouped_date.describe()
grouped_station.describe()
```

```python
grouped_date.aggregate(np.sum)
```
